Todos os dias o nosso sistema atualiza as vendas do dia anterior. O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior.

```python
import pyautogui
import pyperclip
import time

pyautogui.PAUSE = 1

# Passo 1: Entrar no sistema (Uma pasta no google drive)
# pyautogui.hotkey("ctrl", "t")

# Caso tenha que abrir o navegador manualmente
pyautogui.press("win")
pyautogui.write("chrome")
pyautogui.press("enter")

time.sleep(2)
pyautogui.click(x=596, y=424) # Selecionando o usuário

time.sleep(2)
pyautogui.hotkey("ctrl", "t")

pyperclip.copy("https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")

# Site carregando...
time.sleep(5)

# Passo 2: Navegar no sistema e encontrar a base de dados (arquivo na pasta)
pyautogui.click(x=374, y=285, clicks=2)
time.sleep(2)


# Passo 3: Download da base de dados
pyautogui.click(x=382, y=356) # Seleciona o arquivo
pyautogui.click(x=1164, y=175) # Clica nas opções
pyautogui.click(x=970, y=535) # Faz o download da base de dados
time.sleep(7)
pyautogui.click(x=518, y=443)
time.sleep(7)
```

## Vamos agora ler o arquivo baixado para pegar os indicadores

```python
# Passo 4: Calcular os indicadores (faturamento, quantidade de produtos)
import pandas as pd

tabela = pd.read_excel(r"C:/Users/owesl/Downloads/Vendas - Dez.xlsx")
display(tabela)
```

```python
quantidade = tabela["Quantidade"].sum()

faturamento = tabela["Valor Final"].sum()
```

## Entrando no email e enviando os indicadores


```python
# Passo 5: Entrar no Email
pyautogui.hotkey("ctrl", "t") # Abrir nova aba
pyperclip.copy("https://mail.google.com/mail/0/inbox") # Copiando link
pyautogui.hotkey("ctrl", "v") # Colocando o link
pyautogui.press("enter") # Entrando no site

time.sleep(10)

# Passo 6: Mandar um email para a diretoria com os indicadores

# Clicar no botão +
pyautogui.click(x=37, y=186)
time.sleep(2)

# Escrever o destinatário
pyautogui.write("mail@gmail.com")
pyautogui.hotkey("tab")
pyautogui.hotkey("tab")

# Escrever o assunto do Email
# pyautogui.click(x=1234, y=339)
pyperclip.copy("Relatório de Vendas")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

# Escrever o corpo do Email
texto = f"""
Prezados, bom dia!

O faturamento de ontem foi de: R${faturamento:,.2f}
A quantidade de produtos foi de: {quantidade:,}

Abs:
WP
"""
pyperclip.copy(texto)
pyautogui.hotkey("ctrl", "v")


# Enviar o Email
# pyautogui.click(x=845, y=686)
pyautogui.hotkey("ctrl", "enter")
```
